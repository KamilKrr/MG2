### Aufgabe 1

Berechnen Sie die Fläche zwischen den Funktionen $f$ und $g$ im
Intervall zwischen den reellwertigen Schnittpunkten beider
Funktionen:

$$
f(x)=x^2,g(x)=\sqrt{x}
$$

Plotten Sie die beiden Funktionen.

```python
f = x**2
g = sqrt(x)
p1 = plot(f, xlim = (-1,2), ylim = (-1,2), show=False)
p1.extend(plot(g, line_color = 'orange', show=False))
p1.show()
```

$$ 
    \int_{0}^{1} {\left( \sqrt{x} - x^2 \right) dx} 
    = \frac{2}{3}x^{\frac{3}{2}} - \frac{1}{3}x^3 \bigg|_{0}^{1}
    = \frac{1}{3}
$$

```python
i = Integral(g-f,(x,0,1));i
i.doit()
```

Grenzen bestimmen:

$$ 
    \int_{-1}^{3} {\left( 4x - (x^2 + 2x - 3) \right) dx} 
    = \int_{-1}^{3} {\left( - x^2 + 2x + 3 \right) dx}

### Aufgabe 3

Berechnen Sie die Fläche zwischen den Funktionen $f$ und $g$ im
Intervall zwischen den reellwertigen Schnittpunkten beider
Funktionen:

$$
f(x)=\log{x},g(x)=(\log{x})^2
$$

Plotten Sie die beiden Funktionen.

```python
f = ln(x)
g = (ln(x))**2
p1 = plot(f, xlim = (0.5,3), ylim = (-0.5,1.5), show=False)
p1.extend(plot(g, line_color = 'orange', show=False))
p1.show()
```

Grenzen bestimmen:

$$ \log{x} = (\log{x})^2 $$

Substitution:

$$ u = \log{x} $$
$$ u = u^2 $$
$$ u = 0, u = 1 $$
$$ \log{x} = 0: x = 1 $$
$$ \log{x} = 1: x = e $$

$$ 
    \int_{1}^{e} {\left( \log{x} - (\log{x})^{2} \right) dx} 
$$

$$ \int_{1}^{e} {\left( \log{x} \right) dx} = x\log{x}-x \bigg|_{1}^{e} $$

Partielle Integration:
$$ 
    \int_{1}^{e} {\left( (\log{x})^{2} \right) dx} 
    = \int_{1}^{e} {\left( (\log{x})^{2} * 1 \right) dx} 
    = (\log{x})^{2}x - \int_{1}^{e} {\left( \frac{2\log{x}}{x}x \right) dx} \bigg|_{1}^{e}
    = (\log{x})^{2}x - \int_{1}^{e} {\left( 2\log{x} \right) dx} \bigg|_{1}^{e}
    = (\log{x})^{2}x - 2(x\log{x}-x) \bigg|_{1}^{e}
$$

$$ 
    \int_{1}^{e} {\left( \log{x} - (\log{x})^{2} \right) dx} 
    = x\log{x}-x - ((\log{x})^{2}x - 2(x\log{x}-x)) \bigg|_{1}^{e}
    = -(\log{x})^{2}x + 3x\log{x}-3x \bigg|_{1}^{e}
    = -e - (-3) = 3 - e
$$

```python
i = Integral(f-g,(x,1,E));i
i.doit()
```

Dreht man eine Kurve mit $y=f(x)$ um die $x$-Achse, so beträgt das
Volumen des entstehenden Drehkörpers:

$$V=\pi\int_a^b{y^2dx}$$

Die Kurve mit $y=\cosh{x}$ rotiert um die $x$-Achse. Berechnen Sie
das Volumen des Drehkörpers zwischen $a=0$ und $b=2$.

Plotten Sie die Kurve.

```python
f = cosh(x)
p1 = plot(f, xlim = (0, 2), ylim = (-1,5), show=False)
p1.show()
```

$$\int_0^2{(\cosh{x})^2dx}$$

$$\cosh{x} = \frac{e^{x} + e^{-x}}{2}$$
$$\sinh{x} = \frac{e^{x} - e^{-x}}{2}$$

$$\int_0^2{(\frac{e^{x} + e^{-x}}{2})^2dx}$$

$$\int_0^2{\frac{(e^{x} + e^{-x})^2}{4}dx}$$

$$\frac{1}{4}\int_0^2{(e^{x} + e^{-x})^2}dx$$

Binomische Formel:

$$\frac{1}{4}\int_0^2{(e^{x})^2 + 2 + (e^{-x})^2}dx$$

$$\frac{1}{4}\int_0^2{e^{2x} + 2 + e^{-2x}}dx$$

$$\frac{1}{4}(\frac{1}{2}e^{2x} + 2x - \frac{1}{2}e^{-2x}) \bigg|_{0}^{2}$$

$$\frac{x}{2} + \frac{1}{4}(\frac{e^{2x} - e^{-2x}}{2}) \bigg|_{0}^{2}$$

$$\frac{x}{2} + \frac{1}{4}(\sinh{2x}) \bigg|_{0}^{2}$$

Einsetzen:
$$ 7.82247 - 0 $$

$$V=\pi * 7.82247 = 24.5750$$

```python
i = Integral((cosh(x))**2,(x,0,2));i
i.doit()
```

### Aufgabe 8


Die Flächenschwerpunktskoordinaten $x_0,y_0$ einer Fläche $A$
ergeben sich aus den statischen Momenten $M_y,M_x$:

$$A=\int_a^bydx, M_y=\int_a^bxydx, M_x=\frac{1}{2}\int_a^by^2dx, x_0=\frac{M_y}{A}, y_0=\frac{M_x}{A}$$

Berechnen Sie die Flächenschwerpunktskoordinaten für einen Halbkreis
mit $M(0|0)$ und $y\geq0$.

Plotten Sie den Halbkreis.

```python
r = 3
f = sqrt(r**2-x**2)
p1 = plot(f, xlim = (-4, 4), ylim = (-1,4), show=False)
p1.show()
```

Kreisformel:
$$x^2+y^2=r^2$$
Umformen:
$$y = \sqrt{r^2-x^2}$$

Halbkreisfläche:
$$A=\frac{\pi r^2}{2}$$

$$A=\int_a^bydx, M_y=\int_a^bxydx, M_x=\frac{1}{2}\int_a^by^2dx, x_0=\frac{M_y}{A}, y_0=\frac{M_x}{A}$$

$$ M_y=\int_{-r}^{r}xydx$$
$$M_y=\int_{-r}^{r}x\sqrt{r^2-x^2}dx$$

Substitution:
$$z = r^2-x^2 => \frac{dz}{dx} = -2x => dx = -\frac{dz}{2x}$$

$$-\frac{1}{2}\int_{-r}^{r}\sqrt{z}dz$$
$$-\frac{1}{2}\frac{z^\frac{3}{2}}{\frac{3}{2}}dz\bigg|_{-r}^{r}$$
$$-\frac{z^\frac{3}{2}}{3}\bigg|_{-r}^{r}$$

Rücksubstituieren:
$$-\frac{(r^2-x^2))^\frac{3}{2}}{3}\bigg|_{-r}^{r}$$

Da r immer mit einer geraden Potenz vorkommt muss das bestimmte Integral 0 ergeben

$$M_y = 0$$

$$x_0=\frac{M_y}{A} = 0$$
Die x Koordinate liegt immer bei 0 da der Halbkreis entland der Y Achse Symmetrisch ist

$$M_x=\frac{1}{2}\int_{-r}^{r}y^2dx$$

$$M_x=\frac{1}{2}\int_{-r}^{r}r^2-x^2dx$$

$$
    \int_{-r}^{r}r^2-x^2dx
    = r^2\int_{-r}^{r}1-x^2dx
    = r^2x-\frac{x^3}{3}\bigg|_{-r}^{r}
$$
Einsetzen x = r, x = -r und ausrechnen
$$ r^2*r-\frac{r^3}{3}-(-r^3-\frac{-r^3}{3}) $$
$$ r^3 - \frac{r^3}{3}+r^3-\frac{r^3}{3}$$
$$ 2r^3-2\frac{r^3}{3}$$
$$\frac{6r^3}{3}-2\frac{r^3}{3}=\frac{4r^3}{3}$$

$$M_x=\frac{1}{2}\int_{-r}^{r}y^2dx = \frac{1}{2}\frac{4r^3}{3} = \frac{2r^3}{3}$$

$$y_0=\frac{M_x}{A}=\frac{\frac{2r^3}{3}}{\frac{\pi r^2}{2}}=\frac{4r^3}{3\pi r^3}=\frac{4r}{3\pi}$$

$$x_0=0$$
$$y_0=\frac{4r}{3\pi}$$
